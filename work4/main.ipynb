```python
import nltk
import os
import re

import pandas as pd

from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def process(folder):
    lemmatizer = WordNetLemmatizer()

    files = os.listdir(folder)

    documents = []

    for file_name in files:
        file_path = os.path.join(folder, file_name)
        if os.path.isfile(file_path): 
            with open(file_path, 'r') as file:
                text = file.read()

                text = re.sub(r'[^а-яА-ЯёЁa-zA-Z. ]+', '', text)
                text = text.lower()
                # for _1 in range(10, 1, -1):
                #     text = text.replace(' ' * _1, ' ')

                words = nltk.word_tokenize(text)

                without_stop_words = []
                words_lemanized = []
                for word in words:
                    if not(word in stopwords.words("english")) and (word != '.'):
                        without_stop_words.append(word)
                        word_lemanized = lemmatizer.lemmatize(word)
                        words_lemanized.append(word_lemanized)

                text = ''
                for word in words_lemanized:
                    text += word + ' '

                documents.append(text)

    return documents
```

```python
documents = process('txt_sentoken/neg')
```

```python
count_vectorizer = CountVectorizer()

bag_of_words = count_vectorizer.fit_transform(documents)
feature_names = count_vectorizer.get_feature_names_out()
bag_of_words = pd.DataFrame(bag_of_words.toarray(), columns = feature_names)

bag_of_words
```
