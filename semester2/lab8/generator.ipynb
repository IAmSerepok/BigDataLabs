```python
import os
import time

import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import layers
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
generator = tf.keras.models.load_model('models/generator.keras')
discriminator = tf.keras.models.load_model('models/discriminator.keras')
```

```python
N, M = 1, 4

generated_images = [generator(tf.random.normal([1, 100]), training=False) for _ in range(N * M)]

plt.figure(figsize=(12, 6))

for idx, generated_image in enumerate(generated_images):
    plt.subplot(N, M, idx + 1)
    plt.imshow(generated_image[0, :, :, 0], cmap='gray')

plt.show()
```
